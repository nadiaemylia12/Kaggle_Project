#### Remove column with 90% empty

```python
threshold = len(weather_data)*0.1
threshold
```

```python
weather_data.dropna(thresh=threshold, axis=1, inplace=True)
```

```python
print(weather_data.isnull().sum())
```

#### Fill null with appropriate values using mean, median, mode

```python
def impute_median(series):
    return series.fillna(series.median())
```

```python
weather_data.avg_temp = weather_data['avg_temp'].transform(impute_median)
weather_data.min_temp = weather_data['min_temp'].transform(impute_median)
weather_data.max_temp = weather_data['max_temp'].transform(impute_median)
weather_data.max_wind_speed = weather_data['max_wind_speed'].transform(impute_median)
weather_data.most_wind_direction = weather_data['most_wind_direction'].transform(impute_median)
weather_data.avg_relative_humidity = weather_data['avg_relative_humidity'].transform(impute_median)
```

```python
weather_data.isnull().sum()
```

#### Change data type

```python
weather_data['code'] = weather_data['code'].apply(lambda x: str(x))
```

```python
weather_data['date'] = pd.to_datetime(weather_data['date'])
```

#### Add month column

```python
weather_data['month'] = weather_data['date'].dt.month
weather_data.head()
```

#### Visualize average temperature per month

```python
weather_month = pd.DataFrame(weather_data.groupby('month').sum())
weather_month.head()
```

```python
weather_month.sort_values('avg_temp', ascending=False)
```

```python
months = range(1,13)

plt.bar(months, weather_month['avg_temp'])
plt.xticks(months)
plt.xlabel('Month')
plt.ylabel('Average Temperature')
plt.show()
```

#### Visualize average temperature by province

```python
weather_province = pd.DataFrame(weather_data.groupby('province').sum())
weather_province.head()
```

```python
weather_province.sort_values('avg_temp', ascending=False)
```

```python
provinces = [i for i, df in weather_data.groupby('province')]

plt.bar(provinces, weather_province['avg_temp'])
plt.xticks(provinces, rotation=90)
plt.xlabel('Province')
plt.ylabel('Average Temperature')
plt.show()
```

#### Visualize average temperature and precipitation by province

```python
precipitations = weather_data.groupby('province').sum()['precipitation']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(provinces, weather_province['avg_temp'])
ax2.plot(provinces, precipitations, 'g')

ax1.set_xlabel('Province')
ax1.set_ylabel('Average Temperature', color ='b')
ax2.set_ylabel('Precipitation', color='g')
ax1.set_xticklabels(provinces, rotation=90)

plt.show()
```

#### Visualize average temperature and average relative humidity by province

```python
humidities = weather_data.groupby('province').sum()['avg_relative_humidity']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(provinces, weather_province['avg_temp'])
ax2.plot(provinces, humidities, 'g')

ax1.set_xlabel('Province')
ax1.set_ylabel('Average Temperature', color ='b')
ax2.set_ylabel('Average Relative Humidity', color='g')
ax1.set_xticklabels(provinces, rotation=90)

plt.show()
```

#### Visualize minimum and maximum temperature by province

```python
grp = weather_data.groupby('province')
x = grp['min_temp'].agg(np.mean)
y = grp['max_temp'].agg(np.mean)
print(x)
print(y)
```

```python
plt.figure(figsize=(16,5))
plt.plot(x, 'ro')
plt.xticks(rotation=90)
plt.title('Temperature over Province')
plt.ylabel('Average Minimum Temperature')
plt.xlabel('Province')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot(y, 'ro', color='g')
plt.xticks(rotation=90)
plt.title('Temperature over Province')
plt.ylabel('Average Maximum Temperature')
plt.xlabel('Province')
plt.grid()
plt.show()
```
